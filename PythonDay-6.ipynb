```python
# Sum of first 10 natural numbers      #  1+2+3+4.....10  

sum=0
first_num=int(input())
last_num=int(input())

while(first_num <=last_num):
    sum=sum+first_num  # 15
    first_num=first_num+1  #6
print(sum)
```

# FOR LOOP

Syntax:
for variable in sequence:              # Sequence can be list, tuple, string
                Body of for loop       


```python
num=range(12,28,5)  # range(start,stop,step)

# 1 2 3 4 5
for i in num:
    print(i,end=" ")
```

```python
num_range=range(5,-6,-1)
for var_int in num_range:
    print(var_int,end=" ")
```

```python
first_num=int(input())
last_num=int(input())
num=range(first_num,last_num)
sum=0
for i in num:
    sum=sum+i
print(sum)
```

```python
num_list = [4,6,8,10,12,14,16]  # List
for var in num_list:
    print(var,end=" ")
```

```python
num = [4,6,8,10,12,14,16]  # List
for i in range(len(num)):
    print(i,num[i])
```

```python
num = [4,6,8,10,12,14,16]  # List
for i,value in enumerate(num):
    print(i,value)
```
