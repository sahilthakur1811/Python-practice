```python
# Count number of digits
num=int(input())
count=0
if num==0:
    print("1")
else:
    while num!=0:
        count=count+1
        num=num//10
    print(count)
```

```python
# Count of even numbers and odd numbers
start=int(input())
stop=int(input())
num=range(start,stop+1)

countodd=0
counteven=0
for i in num:
    if(i%2==0):
        counteven=counteven+1
    else:
        countodd=countodd+1
print("The count of even numbers is {} and count of odd numbers is {}".format(counteven,countodd))
```

```python
# Number is prime or not prime
num=int(input())
i=2
while i<num:
    if num%i==0:
        print("Not prime")
        break
    i=i+1
else:
    print("Prime Number")
```

```python
# Even Numbers and Odd Numbers

start=int(input())
stop=int(input())

num=range(start,stop+1) # 1 to 6 (excluded)

counteven=0
countodd=0

for i in num:
    if i%2==0:
        counteven=counteven+1 #2
    else:
        countodd=countodd+1    #3
print("Even numbers are", counteven)
print("Odd numbers are ", countodd)

```

```python
n=int(input())
firstHalf=(n+1)//2
secHalf=(n)//2

#firsthalf
currRow=1
while currRow<=firstHalf:
    spaces=1
    while spaces<=(firstHalf-currRow):
        print(' ',end="")
        spaces+=1
    currcol=1
    while currcol<=(2*currRow)-1:
        print("*",end="")
        currcol+=1
    print()
    currRow+=1
#secondhalf
currRow=secHalf
while currRow>=1:
    spaces = 1
    while spaces<=(secHalf-currRow+1):
        print(" ",end="")
        spaces+=1
    currcol=1
    while currcol<=(2*currRow)-1:
        print("*",end="")
        currcol+=1
    print()
    currRow-=1
```
